```python
import torch
import numpy as np
```

```python
sample = "goodmorning"

# Make dictionary
# char_set = list(sample)
char_set = list(set(sample))
# print(sample)
char_dic = {c : i for i, c in enumerate(char_set)}
# print(char_dic)

# Hyper parameters
dic_size = len(char_dic)
hidden_size = dic_size
# print(dic_size)
learning_rate = 0.01
total_epoch = 1000

# Data setting
sample_idx = [char_dic[c] for c in sample]
# print(sample_idx)
x_data = [sample_idx[:-1]]
# print(x_data)
x_one_hot = [np.eye(dic_size)[x] for x in x_data]
# print(x_one_hot)
y_data = [sample_idx[1:]]
# print(y_data)
```

```python
# Transform as torch tensor variables

X = torch.FloatTensor(x_one_hot)
Y = torch.LongTensor(y_data)
# print(X, Y)
```

```python
# Declare RNN
rnn = torch.nn.RNN(dic_size, hidden_size, batch_first=True) # (batch, seq, feature)

# Loss function & Optimizer
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(rnn.parameters(), learning_rate)

# Start training
for i in range(total_epoch+1):
    optimizer.zero_grad()
    outputs, _status = rnn(X)    
#     print(outputs.view(-1, dic_size))
#     print(X)
#     print(Y.view(-1))
    loss = criterion(outputs.view(-1, dic_size), Y.view(-1))
    loss.backward()
    optimizer.step()    
#     print(outputs.data.numpy().argmax(axis=2))
#     break
    
    result = outputs.data.numpy().argmax(axis=2)        
    result_str = ''.join([char_set[c] for c in np.squeeze(result)])    
    print(i, "loss: ", loss.item(), "prediction: ", result, "true Y: ", y_data, "prediction str: ", result_str)
```
